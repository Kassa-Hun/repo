```python
import pandas as df

github_url = 'https://github.com/Kassa-Hun/repo/blob/main/employee.csv'
raw_url = github_url.replace("github.com", "raw.githubusercontent.com").replace("/blob/", "/")

df = df.read_csv(raw_url)
print(df.head())
print(df['MonthlyIncome'].max())
print(df['MonthlyIncome'].min())
print(df['MonthlyIncome'].mean())
print(df['MonthlyIncome'].median())
print(df['MonthlyIncome'].mode().values[0])
print("Range = ", df['MonthlyIncome'].max()-df['MonthlyIncome'].min())
print("Variance = {0:.2f}".format(df['MonthlyIncome'].var()))
print("Std Deviation = {0:.2f}".format(df['MonthlyIncome'].std()))
```

```python
from os import stat
corr_salary_bonus=df['MonthlyIncome'].corr(df['TotalWorkingYears'])
print(corr_salary_bonus)

import scipy.stats as stats
sary_skewness=stats.skew(df['MonthlyIncome'])
print(sary_skewness)

sary_kurstosis=stats.kurtosis(df['MonthlyIncome'])
print(sary_kurstosis)
import numpy as np

salary_percentile = np.percentile(df['MonthlyIncome'],[15,45,65])
salary_quartile=np.percentile(df['MonthlyIncome'],[25,50,75,100])
print("15th percentile = ",salary_percentile[0])
print("45th percentile (Median) = ",salary_percentile[1])
print("65th percentile = ",salary_percentile[2])
print("1st quartile = ",salary_quartile[0])
print("2nd quartile (Median)= ",salary_quartile[1])
print("3rd quartile = ",salary_quartile[2])
print("4th quartile = ",salary_quartile[3])

# Calculate z-scores and assign directly to the DataFrame column
# This ensures the 'z_score' column is present in 'df' before plotting.
df['z_score'] = stats.zscore(df['MonthlyIncome'])

print("z-score of salary", df['z_score'])
import matplotlib.pyplot as plt
df.plot(kind='scatter',x='MonthlyIncome',y='z_score')
plt.show()
```

```python
import pandas as pd

github_url = 'https://github.com/Kassa-Hun/repo/blob/main/matches.xlsx'
raw_url = github_url.replace("github.com", "raw.githubusercontent.com").replace("/blob/", "/")

matches_df = pd.read_excel(raw_url)
total_matches=len(matches_df)
team_wins=len(matches_df[matches_df['winner']=='Mumbai Indians'])
team_losses=len(matches_df[matches_df['winner']!='Mumbai Indians'])
win_prob=team_wins/total_matches
lose_prob=team_losses/total_matches
print("Team probability of wins = {0:.2f}%".format(win_prob*100))
print("Team probability of losses = {0:.2f}%".format(lose_prob*100))
#Set theory
umpire_1=set(matches_df['umpire1'])
umpire_2=set(matches_df['umpire2'])
union_umpire=umpire_1.union(umpire_2) # A u B
print("Union set umpires = ",union_umpire)
interception_umpire=umpire_1.intersection(umpire_2) #A n B
print("Interception umpires = ",interception_umpire)
difference_umpire=umpire_1.difference(umpire_2) #A-B
print("Difference umpires = ",difference_umpire)

#Conditional probability
toss_wins=len(matches_df[matches_df['toss_winner']=='Mumbai Indians'])
matches_win_given_toss=len(matches_df[(matches_df['toss_winner']=='Mumbai Indians') & (matches_df['winner']=='Mumbai Indians')])
conditional_prob=matches_win_given_toss/ toss_wins
print("Conditional probability of win given toss= {0:.2f}%".format(conditional_prob*100))

#Conditional probability
toss_field=len(matches_df[matches_df['toss_decision']=='field'])
matches_win_given_field=len(matches_df[(matches_df['toss_decision']=='field') & (matches_df['winner']=='Mumbai Indians')])
conditional_probability=matches_win_given_field/ toss_field
print("Conditional probability of win given field= {0:.2f}%".format(conditional_probability*100))
```
